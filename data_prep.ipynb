### Drop Unimportant Coulmns & Change Columns Name 

```python
df1 = df1[["Subject","AGE","PTGENDER","PTEDUCAT","MMSE","DX.bl"]]

df2 = df2[["ID","Age","M/F","Educ","MMSE","eTIV","nWBV","CDR"]]

df3 = df3[["Subject ID","Age","M/F","EDUC","MMSE","eTIV","nWBV","CDR"]]

kaggle = kaggle[['PatientID', 'Age', 'Gender', 'EducationLevel', 'MMSE', 'Diagnosis']]
```

```python
df1.columns = ["ID","Age","Gender","Educ","MMSE","Target"]

df2.columns = ["ID","Age","Gender","Educ","MMSE","eTIV","nWBV","Target"]

df3.columns = ["ID","Age","Gender","Educ","MMSE","eTIV","nWBV","Target"]

kaggle.columns = ["ID","Age","Gender","Educ","MMSE","Target"]
```

### Merging All Datasets

```python
df4 = pd.concat([df1, df2, df3, kaggle], ignore_index=True)
df4.head()
```

```python
df4.info()
```

### Handling The Merged Dataset

```python
df4.value_counts('Target')
```

```python
df4['Target'] = df4['Target'].replace({0: 0, 0.5: 1, 1: 1, 2: 1})
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df4[['eTIV', 'nWBV']] = imputer.fit_transform(df4[['eTIV', 'nWBV']])
df4.info()
```

```python
df4.dropna(subset=['Target'], inplace=True)
df4.info()
```

###  Standardization The Dataset

```python
def harmonize_education(value):
    if pd.isna(value): return np.nan
    if isinstance(value, (int, float)):
        if value > 6: # Years of Education (e.g., 16 years)
            if value < 12: return 0
            if value == 12: return 1
            if 13 <= value <= 15: return 2
            if value == 16: return 3
            if value > 16: return 4
        elif 1 <= value <= 5: # OASIS-1 Scale (1-5)
            return int(value - 1)
        elif 0 <= value <= 4: # Already standardized
            return int(value)
    return np.nan
```

```python
df4['Educ'] = df4['Educ'].apply(harmonize_education)
df4['Educ'].value_counts()
```

```python
df4.dropna(subset=['MMSE'], inplace=True)
df4.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
continuous_cols = ['Age', 'MMSE', 'eTIV', 'nWBV']
df4[continuous_cols] = scaler.fit_transform(df4[continuous_cols])
```
