```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('ADNI_Training_Q3_APOE_CollectionADNI1Complete 1Yr 1.5T_July22.2014.csv')
df1.head()
```

```python
df1['PTGENDER'] = df1['PTGENDER'].replace({'Male': 1, 'Female': 0})

df1['DX.bl'] = df1['DX.bl'].replace({"CN": 0, "LMCI": 0.5, "AD": 1})


```

```python
df3["M/F"] = df3["M/F"].replace({'M': 1, 'F': 0})
```
